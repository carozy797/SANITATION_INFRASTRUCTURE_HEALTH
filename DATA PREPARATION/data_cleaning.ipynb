```python

def add_city_with_two(commuity_iD, year, data):
    df_2011_health_city1_ID1 =  data[data["communityID"] == commuity_iD[0]]
    df_2011_health_city1_ID2 =  data[data["communityID"] == commuity_iD[1]]
    df_2011_health_city1_ID3 =  data[data["communityID"] ==commuity_iD[2]]
    df_2011_health_city1_ID4 =  data[data["communityID"] == commuity_iD[3]] 
    df_2011_health_city1 = pd.concat([df_2011_health_city1_ID1,df_2011_health_city1_ID2, df_2011_health_city1_ID3, df_2011_health_city1_ID4], ignore_index=True)
    print(df_2011_health_city1)
    df_2011_health_status = pd.DataFrame()
    df_2011_health_status["Year"] = [year for i in range(len(df_2011_health_city1))]
    df_2011_health_status["ID"] = df_2011_health_city1["ID"]
    df_2011_health_status["HouseHoldID"] = df_2011_health_city1["householdID"]
    df_2011_health_status["communityID"] = df_2011_health_city1["communityID"]
    df_2011_health_status["self_com_on_health"] = df_2011_health_city1["da002"]
    df_2011_health_status["hyp"] = df_2011_health_city1["da007_1_"]
    df_2011_health_status["mem_rel_dis"] = df_2011_health_city1["da007_12_"]
    df_2011_health_status["rheumatism"] = df_2011_health_city1["da007_13_"]
    df_2011_health_status["lung_dis"] = df_2011_health_city1["da007_5_"]
    df_2011_health_status["liver_dis"] = df_2011_health_city1["da007_6_"]
    df_2011_health_status["stomach_dis"] = df_2011_health_city1["da007_10_"]
    return df_2011_health_status
```

```python
mapping = {'Yes': 1, 'No': 0}
mapping_comment_on_health = {
'Fair': 3,          
'Poor': 4,         
'Good': 2,          
'Very good':1,   
'Very poor': 5, 
}

mapping_cooking_fuel ={ 
'Natural Gas': 5,                 
'Electric': 2,                       
'Coal': 6,                        
'Liquefied Petroleum Gas': 3,         
'Crop residue/Wood burning'  : 1,                         
'Marsh gas': 4,                   
}

mapping_main_toilet_type = {
'Inside toilet with water': 5,                
'Open-air'    : 1,                           
'Outside public toilet without water': 2,     
'Outside toilet with water': 3,          
'Inside toilet without water': 4,         
}

mapping_waste_management = {
'Moved away by truck': 5,        
'Buried in this village': 3,     
'Do not manage' : 1,            
'Burn away' : 2,                   
}

mapping_road_type = {
'Paved road': 3,                    
'Pathway/Dirt/unpaved road': 2,     
'Sand-stone road': 1,                
}
     
mapping_inhouse_shower_Bath = {
'Water heater installed by the household' : 2,    
'No': 1,                                        
'Hot water provided' : 3,                     
}
                                           
data_new = above_60
sewage_encode = data_new["sewage_sys"].map(mapping)
stomach_dis_encode = data_new["stomach_dis"].map(mapping)
liver_dis_encode = data_new["liver_dis"].map(mapping)
lung_dis_encode = data_new["lung_dis"].map(mapping)
self_com_on_health_encode =  data_new["self_com_on_health"].map(mapping_comment_on_health)
rheumatism_encode =  data_new["rheumatism"].map(mapping)
cooking_fuel_source_encode = data_new["cooking_fuel_source"].map(mapping_cooking_fuel)
mem_rel_encode =  data_new["mem_rel_dis"].map(mapping)
road_type_encode =  data_new["road_type"].map(mapping_road_type)
waste_management_encode =  data_new["waste_management"].map(mapping_waste_management)
main_toilet_type_encode =  data_new["main_toilet_type"].map(mapping_main_toilet_type)
inhouse_shower_Bath_encode =  data_new["inhouse_shower_Bath"].map(mapping_inhouse_shower_Bath)
hyp_encode = data_new["hyp"].map(mapping)
has_running_water_encode = data_new["has_running_water"].map(mapping)
```

```python
# Sanitation
coded_data = pd.DataFrame()
coded_data["Year"] = above_60["Year"]
coded_data["province"] = above_60["province"]
coded_data["city"] = above_60["city"]
coded_data["communityID"] = above_60["communityID"]
coded_data["road_type"] = road_type_encode
coded_data["sewage_sys"] = sewage_encode
coded_data["main_toilet_type"] = main_toilet_type_encode
coded_data["cooking_fuel_source"] = cooking_fuel_source_encode
coded_data["has_running_water"] = has_running_water_encode
coded_data["inhouse_shower_Bath"] = inhouse_shower_Bath_encode
coded_data["waste_management"] = waste_management_encode
# Health
coded_data["stomach_dis"] = stomach_dis_encode
coded_data["liver_dis"] = liver_dis_encode
coded_data["lung_dis"] = lung_dis_encode
coded_data["rheumatism"] = rheumatism_encode
coded_data["mem_rel_dis"] = mem_rel_encode
coded_data["hyp"] = hyp_encode
coded_data["self_com_on_health"] = self_com_on_health_encode
coded_data["health_status"] = above_60["health_status"]

coded_data.isna().sum()
coded_data
coded_data.to_csv("./coded_data.csv", index=False)
```
