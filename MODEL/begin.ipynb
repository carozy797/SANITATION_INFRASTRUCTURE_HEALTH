```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder

# Initialize the OneHotEncoder
encoder = OneHotEncoder()
used_data = data
# Define the columns to be one-hot encoded
columns_to_encode = ['has_running_water','inhouse_shower_Bath','cooking_fuel_source', 'toilet_type_1', 'hyp', 'diab','mem_rel_dis', 'asthma', 'rheumatism', 'cancer', 'lund_dis','liver_dis', 'stomach_dis', 'main_toilet_type', 'sewage_sys','road_type']

# One-hot encode each column and store the results in separate DataFrames
encoded_dfs = {}
for column in columns_to_encode:
    encoded_data = encoder.fit_transform(used_data[[column]])
    encoded_df = pd.DataFrame(encoded_data.toarray(), columns=encoder.get_feature_names_out([column]))
    encoded_dfs[column] = encoded_df

# Concatenate the encoded DataFrames along the columns (axis=1)
combined_encoded_df = pd.concat(encoded_dfs.values(), axis=1)

# Drop the original columns from the original data
# used_data = used_data.drop(columns=columns_to_encode)
used_data = pd.DataFrame()
# Concatenate the original data with the combined encoded DataFrame
final_df = pd.concat([used_data, combined_encoded_df], axis=1)

# Check the correlation
# correlation_matrix = final_df.corr()
# correlation_matrix
```
