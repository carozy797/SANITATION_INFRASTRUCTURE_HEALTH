Chi-Square Between Variables

```python
import pandas as pd
from scipy.stats import chi2_contingency

# Your variable lists

variab = ['hyp', 'mem_rel_dis', 'rheumatism', 'lung_dis', 'liver_dis', 'stomach_dis', 'self_com_on_health']
variables = ['has_running_water', 'inhouse_shower_Bath', 'cooking_fuel_source', 'main_toilet_type', 'sewage_sys', 'road_type', 'waste_management']

# Create an empty list to store the results
results = []

# Perform the chi-square tests and store the results
for var in variab:
    crosstab = pd.crosstab(data['cooking_fuel_source'], data[var])
    chisquare, p_value, dof, exv = chi2_contingency(crosstab)
    
    # Store the results in a dictionary
    result_dict = {
        'Independent Variable': 'cooking_fuel_source',
        'Target Variable': var,
        'Chi-Square': chisquare,
        'P-Value': p_value,
     
    }
    results.append(result_dict)

# Create a dataframe from the results list
result_df = pd.DataFrame(results)

# Print the dataframe
result_df
# result_df.to_csv("./cooking_fuel_source_results.csv")
```
