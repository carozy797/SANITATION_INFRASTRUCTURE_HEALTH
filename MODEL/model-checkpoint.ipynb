The Provinces and other detailes across the years

```python
import pandas as pd

# Assuming you have a DataFrame named 'community_profile' with columns: 'Year', 'province', 'city', 'communityID'

# Group by 'Year' and then within each year, calculate value counts for 'province', 'city', and 'communityID'
result_series = community_profile.groupby(['Year', 'province', 'city', 'communityID']).size()

# Convert the Series to a DataFrame
result_df = result_series.reset_index(name='count')

result_df
```

```python
dt = pd.read_excel("../DATA PREPARATION/health_combined.xlsx",)
```

Getting the English Names

```python
import pandas as pd

# List of community IDs
community_ids = [540541, 940042, 1040041, 1082592, 1940372, 140311, 896763, 1040043, 1051022, 146431,
                 1940373, 104462, 563463, 1051021, 140511, 104461, 1051023, 563021, 563462, 140313,
                 940041, 1053811, 1053813, 1040042, 1940371, 140512, 563022, 1082593, 896762, 3452333,
                 940043, 146432, 3477633]

# Filter rows based on community IDs
filtered_data = dt[dt['communityID'].isin(community_ids)]

# Display the filtered data
filtered_data[['communityID', 'Province_y', 'City_y']]
```

```python

for index, row in result_df.iterrows():
    community_id = row['communityID']
    
    # Find the corresponding row in the 'filtered_data' DataFrame
    filtered_row = filtered_data[filtered_data['communityID'] == community_id]
    
    # Update the province and city values in the 'result' DataFrame
    if not filtered_row.empty:
        result_df.at[index, 'province'] = filtered_row.iloc[0]['Province_y']
        result_df.at[index, 'city'] = filtered_row.iloc[0]['City_y']

# Store the updated 'result' DataFrame
result_df.to_csv('output.csv', index=False)
```

```python
# testing the significance of correlation between health score and sanitation_infrastructure 
```

```python
import scipy.stats as stats 
x = df_coded['Health_score']
y = df_coded['sanitation_infrastructure']
correlation_matrix, p_value = stats.pearsonr(x,y)
print(f'Correlation_matrix between health score and sanitation infrastructure is: ', correlation_matrix, 'p_value:', p_value)
```
