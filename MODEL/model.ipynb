```python
import pandas as pd
from scipy.stats import chi2_contingency

# Your variable lists

variab = ['hyp', 'mem_rel_dis', 'rheumatism', 'lund_dis', 'liver_dis', 'stomach_dis']
variables = ['communityID', 'province', 'city', 'urban_nbs', 'age_cal', 'has_running_water', 'inhouse_shower_Bath', 'cooking_fuel_source', 'num_toilet', 'toilet_type_1', 'hyp', 'diab', 'mem_rel_dis', 'asthma', 'rheumatism', 'cancer', 'lund_dis', 'liver_dis', 'stomach_dis', 'main_toilet_type', 'sewage_sys', 'road_type']

# Create an empty list to store the results
results = []

# Perform the chi-square tests and store the results
for var in variab:
    crosstab = pd.crosstab(data['cooking_fuel_source'], data[var])
    chisquare, p_value, dof, exv = chi2_contingency(crosstab)
    
    # Store the results in a dictionary
    result_dict = {
        'Independent Variable': 'cooking_fuel_source',
        'Target Variable': var,
        'Chi-Square': chisquare,
        'P-Value': p_value,
     
    }
    results.append(result_dict)

# Create a dataframe from the results list
result_df = pd.DataFrame(results)

# Print the dataframe
result_df
# result_df.to_csv("./cooking_fuel_source_results.csv")
```

Regression Model

```python
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, mean_absolute_error
import numpy as np
```

```python
y = df_coded["Health_score"]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.15, random_state=1)
lr = LinearRegression()
model = lr.fit(X_train,y_train)
health_score_pred = model.predict(X_test)
error = np.sqrt(mean_squared_error(health_score_pred, y_test))
print(f"rmse: {error}")
```

Random Forest Regressor

```python
y = df_coded["Health_score"]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.15, random_state=1)
rf = RandomForestRegressor(random_state = 42)
model_rf = rf.fit(X_train,y_train)
health_score_pred = model_rf.predict(X_test)
error_rf = np.sqrt(mean_squared_error(y_test, health_score_pred))
print(f"rmse: {error_rf}")
```

The LR has the least error in this case... We want to know the feature importance of that model

```python

# Get the feature importances
feature_importance = model_rf.feature_importances_

# Create a DataFrame to show the feature importance scores along with their corresponding feature names
feature_importance_df = pd.DataFrame({'Feature': X.columns, 'Importance': feature_importance})

# Sort the DataFrame by importance in descending order
feature_importance_df = feature_importance_df.sort_values(by='Importance', ascending=False)

# Print or visualize the feature importance scores
print(feature_importance_df)

# Optional: Plot a horizontal bar chart to visualize feature importance

plt.figure(figsize=(10, 6))
plt.barh(feature_importance_df['Feature'], feature_importance_df['Importance'], color='g')
plt.xlabel('Importance Score')
plt.ylabel('Feature')
plt.title('Feature Importance of Linear Regression Model')
plt.show()
```
