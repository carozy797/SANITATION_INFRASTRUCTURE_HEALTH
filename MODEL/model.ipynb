cramers_v

```python
import numpy as np

def cramers_v(confusion_matrix):
    chi2 = chi2_contingency(confusion_matrix)[0]
    n = confusion_matrix.sum()
    phi2 = chi2 / n
    r, k = confusion_matrix.shape
    phi2corr = max(0, phi2 - ((k - 1) * (r - 1)) / (n - 1))
    rcorr = r - ((r - 1) ** 2) / (n - 1)
    kcorr = k - ((k - 1) ** 2) / (n - 1)
    return np.sqrt(phi2corr / min((kcorr - 1), (rcorr - 1)))
```

```python
import pandas as pd
from scipy.stats import chi2_contingency

# Load your dataset into a DataFrame
# data = pd.read_csv('data.csv')

# List of sanitation variables and health variables
health_vars = ['hyp', 'mem_rel_dis', 'rheumatism', 'lung_dis', 'liver_dis', 'stomach_dis', 'self_com_on_health']
sanitation_vars = ['has_running_water', 'inhouse_shower_Bath', 'cooking_fuel_source', 'main_toilet_type', 'sewage_sys', 'road_type', 'waste_management']
# Create a dictionary to store aggregate Cramér's V values for each sanitation variable
aggregate_cramers = {}

# Loop through sanitation and health variables to calculate associations
for sanitation_var in sanitation_vars:
    aggregate_cramer = 0
    
    for health_var in health_vars:
        contingency_table = pd.crosstab(data[sanitation_var], data[health_var])
        cramer_v = cramers_v(contingency_table.values)
        aggregate_cramer += cramer_v
        
    # Calculate the mean Cramér's V across all health variables
    mean_cramer = aggregate_cramer / len(health_vars)
    
    # Store the mean Cramér's V in the dictionary
    aggregate_cramers[sanitation_var] = mean_cramer

# Display the mean Cramér's V values for all sanitation variables
for sanitation_var, mean_cramer in aggregate_cramers.items():
    print(f"Sanitation variable: {sanitation_var}")
    print(f"Mean Cramér's V: {mean_cramer:.4f}")
    print("=" * 20)

# Identify the sanitation variable with the highest mean Cramér's V
most_contributing_variable = max(aggregate_cramers, key=aggregate_cramers.get)

print("Most contributing sanitation variable to health variables combined:", most_contributing_variable)
print("Mean Cramér's V:", aggregate_cramers[most_contributing_variable])
```

```python
y = df_coded["Health_score"]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.15, random_state=1)
rf = RandomForestRegressor(random_state = 42)
model_rf = rf.fit(X_train,y_train)
health_score_pred = model_rf.predict(X_test)
error_rf = np.sqrt(mean_squared_error(y_test, health_score_pred))
print(f"rmse: {error_rf}")
```
