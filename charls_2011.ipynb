```python

def add_city_with_two(commuity_iD, year, data):
    df_2011_health_city1_ID1 =  data[data["communityID"] == commuity_iD[0]]
    df_2011_health_city1_ID2 =  data[data["communityID"] == commuity_iD[1]]
    df_2011_health_city1_ID3 =  data[data["communityID"] ==commuity_iD[2]]
    # # df_2011_health_city1_ID4 =  df_2011_health[df_2011_health["communityID"] == "0140512"] 
    df_2011_health_city1 = pd.concat([df_2011_health_city1_ID1,df_2011_health_city1_ID2, df_2011_health_city1_ID3 ], ignore_index=True)
    print(df_2011_health_city1)
    df_2011_health_status = pd.DataFrame()
    df_2011_health_status["Year"] = [year for i in range(len(df_2011_health_city1))]
    df_2011_health_status["ID"] = df_2011_health_city1["ID"]
    df_2011_health_status["HouseHoldID"] = df_2011_health_city1["householdID"]
    df_2011_health_status["communityID"] = df_2011_health_city1["communityID"]
    df_2011_health_status["self_com_on_health"] = df_2011_health_city1["da002"]
    df_2011_health_status["hyp"] = df_2011_health_city1["da007_1_"]
    df_2011_health_status["diab"] = df_2011_health_city1["da007_3_"]
    df_2011_health_status["mem_rel_dis"] = df_2011_health_city1["da007_12_"]
    df_2011_health_status["asthma"] = df_2011_health_city1["da007_14_"]
    df_2011_health_status["rheumatism"] = df_2011_health_city1["da007_13_"]
    df_2011_health_status["cancer"] = df_2011_health_city1["da007_4_"]
    df_2011_health_status["lund_dis"] = df_2011_health_city1["da007_5_"]
    df_2011_health_status["liver_dis"] = df_2011_health_city1["da007_6_"]
    df_2011_health_status["stomach_dis"] = df_2011_health_city1["da007_10_"]
    # df_2011_health_status
    return df_2011_health_status
```

```python
def add_city_with_two_housing(commuity_iD, year, data):
    df_2011_housing_xstics_IDI1= data[data["communityID"] == commuity_iD[0]]
    df_2011_housing_xstics_IDI2 = data[data["communityID"] == commuity_iD[1]]
    df_2011_housing_xstics_IDI3 = data[data["communityID"] == commuity_iD[2]]
    # df_2011_housing_xstics_IDI4 = df_2011_housing_xstics[df_2011_housing_xstics["communityID"] == "0140512"]
    df_2011_housing_xstics_city1 = pd.concat([df_2011_housing_xstics_IDI1,df_2011_housing_xstics_IDI2, df_2011_housing_xstics_IDI3 ], ignore_index=True)
    df_2011_housing_info  = pd.DataFrame()
    df_2011_housing_info["Year"] = [year for i in range(len(df_2011_housing_xstics_city1))]
    # df_2011_housing_info["ID"] = df_2011_housing_xstics_city1["ID"]
    df_2011_housing_info["HouseHoldID"] = df_2011_housing_xstics_city1["householdID"]
    df_2011_housing_info["communityID"] = df_2011_housing_xstics_city1["communityID"]
    df_2011_housing_info["has_running_water"] = df_2011_housing_xstics_city1["i017"]
    df_2011_housing_info["inhouse_shower_Bath"] = df_2011_housing_xstics_city1["i018"] 
    df_2011_housing_info["cooking_fuel_source"] = df_2011_housing_xstics_city1["i022"] 
    df_2011_housing_info["num_toilet"] = df_2011_housing_xstics_city1["i012_3"] 
    df_2011_housing_info["toilet_type_1"] = df_2011_housing_xstics_city1["i014s1"] 
    df_2011_housing_info["toilet_type_2"] = df_2011_housing_xstics_city1["i014s2"] 
    return df_2011_housing_info
```
